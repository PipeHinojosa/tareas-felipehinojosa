```python
# Solicitar datos al usuario
edad = int(input("Ingresa tu edad: "))
le_gusta_musica = input("¿Te gusta la música? (si/no): ").lower()
le_gusta_bailar = input("¿Te gusta bailar? (si/no): ").lower()

# Convertir respuestas a booleanos
musica = le_gusta_musica == "si"
bailar = le_gusta_bailar == "si"

# Evaluar condiciones
if edad >= 18 and musica and bailar:
    print("Eres una estrella de la fiesta")
elif musica and not bailar:
    print("Te gusta la música, pero no tanto bailar")
else:
    print("Tienes tu propio estilo")
```
